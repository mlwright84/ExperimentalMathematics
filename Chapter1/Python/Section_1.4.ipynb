## A sum of two arctangent values
We first consider the formula for $\pi$ from equation (1.7) in the text:

$$ \frac{\pi}{4} = \arctan\left(\frac{1}{2}\right) + \arctan\left(\frac{1}{3}\right) $$

Converting this formula to a series, we obtain equation (1.8) in the text:

$$ \pi=4\sum_{n=1}^{\infty} \frac{(-1)^{n+1}}{2n-1}\left( \frac{1}{2^{2n-1}}+\frac{1}{3^{2n-1}}\right) $$

We use a loop with an accumulator to add up partial sums of this series. We will put our code for this in a function for convenient use later. Our function requires one parameter, $m$, the number of terms to add up.

```python
def piSum(m):
    acc = 0.0 # initialize the accumulator as a local variable

    for n in range(1, m + 1):
        term1 = 1 / (2**(2*n - 1))
        term2 = 1 / (3**(2*n - 1))
        acc += ((-1)**(n + 1) / (2*n - 1)) * (term1 + term2)

    return 4 * acc # return value
```

Since we wrote our code using standard floating-point numbers, the function returns a decimal approximation directly. For example, here is the result with ten terms of the sum:

```python
val = piSum(10)
val
```

Let's compute many approximations. We'll use a list comprehension to compute a list of results.

```python
maxTerms = 10
piValues = [piSum(n) for n in range(1, maxTerms + 1)]
piValues
```

Now we compute the accuracy of each approximation, using the $\log_{10}$ method as in the text.

```python
import math

accuracy = []
for val in piValues:
    diff = abs(math.pi - val)
    if diff == 0:
        accuracy.append(float('inf'))
    else:
        accuracy.append(-math.log10(diff))

accuracy
```

We can print our findings in a table, similar to Table 1.5 in the text.

```python
import pandas as pd

df = pd.DataFrame({
    'm': range(1, maxTerms + 1),
    'approximation': piValues,
    'accuracy': [round(acc, 5) for acc in accuracy]
})
df
```

Let's plot the accuracy as a function of the number of terms.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.plot(range(1, maxTerms + 1), accuracy, marker='o', color='tab:blue')
plt.xlabel('number of terms')
plt.ylabel('accuracy')
plt.grid(True)
plt.show()
```

Now we will draw the previous plot together with the line through its first and last points.

```python
plt.figure(figsize=(6, 4))

# Plot the accuracy points
plt.plot(range(1, maxTerms + 1), accuracy, marker='o', color='tab:blue', label='Accuracy')

# Plot the line through the first and last points
plt.plot([1, maxTerms], [accuracy[0], accuracy[-1]], color='tab:orange', linewidth=2, label='Line')

plt.xlabel('number of terms')
plt.ylabel('accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
