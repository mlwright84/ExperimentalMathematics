## Dart Board Method
We will implement Algorithm 1.28, which computes an approximation of $\pi$ using a circular dartboard inscribed in a square sampling region. As described in the text, it suffices to randomly select points $(x, y)$ with $0 \le x \le 1$ and $0 \le y \le 1$.

```python
import random
random.random()
```

Python's `random.random()` function returns random values uniformly selected between 0 and 1. Note that you will almost certainly get a new value each time you evaluate the code cell above.

Here is an implementation of Algorithm 1.28 in the text:

```python
def dartBoardPi(m):
    # initialize an accumulator
    acc = 0

    # main loop
    for _ in range(m):
        x = random.random()
        y = random.random()
        r = x**2 + y**2
        if r < 1:
            acc += 1

    # return value
    return 4 * acc / m
```

Run the module and confirm that it produces a value close to $\pi$.

```python
dartBoardPi(100000)
```

To get a sense for the randomness in this approximation method, as well as how the approximation depends on the number of darts, we will run our approximation several times for each number of darts $m \in \{100, 1000, 10000, 100000\}$. Suppose we want to see how variable the resulting approximation is as we repeat the calculation. Here's a table of 10 repetitions for M=100, 1000, 10000, and 100000.

```python
result = [[dartBoardPi(m) for m in [100, 1000, 10000, 100000]] for _ in range(10)]
result
```

We can use `pandas.DataFrame` to see the results more clearly. This produces a table similar to Table 1.8 in the text.

```python
import pandas as pd
df = pd.DataFrame(result, columns=["m=100", "m=1000", "m=10000", "m=100000"])
df
```

We can also draw the points (colored according to being a hit or miss) along with the circumscribed square and the unit circle. First, let's generate the points.

```python
m = 1000 # number of darts
acc = 0 # accumulator
hitList = [] # locations of darts inside the circle
missList = [] # locations of darts outside the circle

# main loop
for _ in range(m):
    # generate random values in [-1, 1]
    x = random.uniform(-1, 1)
    y = random.uniform(-1, 1)
    r = x**2 + y**2

    # check whether the dart lies inside the circle
    if r < 1:
        acc += 1
        hitList.append((x, y))
    else:
        missList.append((x, y))

# print results
print("darts inside circle:", len(hitList))
print("darts outside circle:", len(missList))
print("approximation of pi:", 4 * acc / m)
```

Now we can plot them:

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 6))

# Draw square and circle
ax.add_patch(plt.Rectangle((-1, -1), 2, 2, fill=False, edgecolor='black'))
ax.add_patch(plt.Circle((0, 0), 1, fill=False, edgecolor='black'))

# Plot points
if hitList:
    hx, hy = zip(*hitList)
    ax.scatter(hx, hy, color='red', s=10)
if missList:
    mx, my = zip(*missList)
    ax.scatter(mx, my, color='blue', s=10)

ax.set_aspect('equal', 'box')
ax.set_xlim([-1.1, 1.1])
ax.set_ylim([-1.1, 1.1])
plt.show()
```

## Sampling Dangers
Suppose we attempt to approximate $\pi$ by sampling from a circle and keeping the values inside of the inscribed square, as described in the text. The following code implements Algorithm 1.29.

```python
import math

def incorrectDartBoardPi(m):
    acc = 0
    for _ in range(m):
        r = random.random()
        theta = random.uniform(0, 2 * math.pi)
        x = r * math.cos(theta)
        y = r * math.sin(theta)

        if (-math.sqrt(2)/2 <= x <= math.sqrt(2)/2) and (-math.sqrt(2)/2 <= y <= math.sqrt(2)/2):
            acc += 1

    if acc == 0: return float('inf')
    return 2 * m / acc
```

This function does not appear to return a value that is close to $\pi$.

```python
incorrectDartBoardPi(10000)
```

Let's make a table of results using several values of $m$.

```python
result_incorrect = [[incorrectDartBoardPi(m) for m in [10, 100, 1000, 10000, 100000]] for _ in range(10)]
result_incorrect
```
