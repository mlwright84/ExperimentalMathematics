## Approximating $\pi$ with inscribed $2^n$-gons
This is an implementation of Pseudocode 1.4 in Section 1.2.
Set the number of iterations:

```python
nMax = 10
nMax
```

Set the initial side length of the inscribed square. We will use standard floating-point arithmetic for speed.

```python
import math

sideLength = math.sqrt(2)
sideLength
```

Initialize a list to store the approximations $\pi_n$:

```python
piValues = [0.0] * (nMax + 1)
```

Include our initial approximation $\pi_2$ in the list:

```python
piValues[2] = 2 * sideLength
```

Use a loop to compute $\pi_n$ for $n = 3, 4, \dots, nMax$.
The following code uses a standard Python `for` loop. Note the recurring use of the variable `sideLength`.

```python
for n in range(3, nMax + 1):
    # Compute a and b values
    a = math.sqrt(1 - (sideLength / 2)**2)
    b = 1 - a

    # Compute the new side length and perimeter
    sideLength = math.sqrt(b**2 + (sideLength / 2)**2)
    perimeter = (2**n) * sideLength

    # Compute the approximation to pi
    piValues[n] = perimeter / 2
```

Now we can print the results.

```python
piValues
```

The negative log of the difference is approximately the number of correct digits in our approximation. (See Pseudocode 1.9 and adjacent discussion in the text.)

```python
accuracy = []
for val in piValues:
    if val == 0:
        accuracy.append(0.0)
    else:
        diff = abs(math.pi - val)
        if diff == 0:
            accuracy.append(float('inf'))
        else:
            accuracy.append(-math.log10(diff))

accuracy
```

We can plot the accuracy as a function of $n$.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.plot(range(2, nMax + 1), accuracy[2:], marker='o')
plt.xlabel('n')
plt.ylabel('accuracy')
plt.title('Accuracy of Archimedes Method')
plt.grid(True)
plt.show()
```

## Functions and Runtimes
In programming, a function allows us to give a name to a block of code, which we can call whenever we want to run the code. Here, we define the function `approxPi` from Pseudocode 1.14 in the text. This function requires one parameter, `nMax`.

```python
def approxPi(nMax):
    sideLength = math.sqrt(2)
    piVals = [0.0] * (nMax + 1)
    piVals[2] = 2 * sideLength

    for n in range(3, nMax + 1):
        a = math.sqrt(1 - (sideLength / 2)**2)
        b = 1 - a
        sideLength = math.sqrt(b**2 + (sideLength / 2)**2)
        perimeter = (2**n) * sideLength
        piVals[n] = perimeter / 2

    return piVals[nMax]
```

Next we will measure the runtime of our function. Note that runtime measurements may be inaccurate if the calculation is too fast. We can run the function many times or use larger values of $n$ to get a more accurate measurement of how execution time scales.

```python
import time

def timeIt(nMax):
    start_time = time.time()
    # For standard floats, Archimedes is extremely fast.
    # We loop it 10,000 times just to get a measurable timing in Python.
    for _ in range(10000):
        approxPi(nMax)
    return time.time() - start_time

# Time for nMax = 20 (ran 10,000 times)
timeIt(20)
```

To perform a runtime analysis, let's create a table that tracks the execution time for increasingly large values of $nMax$.

```python
import pandas as pd

test_values = [10, 50, 100, 200, 300, 400, 500]
runtimes = [timeIt(n) for n in test_values]

df_runtime = pd.DataFrame({
    'nMax': test_values,
    'Runtime (seconds per 10k runs)': runtimes
})
df_runtime
```

As expected, using machine precision (standard floating-point arithmetic), the runtime of the Archimedes iterative loop scales roughly linearly with the number of iterations $n$.

```python
plt.figure(figsize=(6, 4))
plt.plot(test_values, runtimes, marker='s', color='red')
plt.xlabel('Number of Iterations (nMax)')
plt.ylabel('Runtime (seconds)')
plt.title('Runtime Scaling of approxPi')
plt.grid(True)
plt.show()
```
